```python
def play(freq, sec=1):
    framerate = 44100  # Sampling rate
    t = np.linspace(0, sec, framerate * sec)  # Setup time values
    data = np.sin(2 * np.pi * freq * t) #Sin function formula
    return Audio(data, rate=framerate, autoplay=True) #Play the generated sound
```

```python
play(400) #Play a sound of 400 Hz for a second
```
